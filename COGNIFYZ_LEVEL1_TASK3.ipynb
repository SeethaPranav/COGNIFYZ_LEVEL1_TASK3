## Check the unique values in 'Price range'

```python
print("Unique Price Ranges:", data['Price range'].unique())
```

## Count how many restaurants are in each price range

```python
price_counts = data['Price range'].value_counts().sort_index()

print("Restaurant counts per price range:")
print(price_counts)
```

## Create a Bar Chart

```python
sns.barplot(x=price_counts.index, y=price_counts.values, palette='viridis')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range (1 = Cheapest, 4 = Most Expensive)')
plt.ylabel('Number of Restaurants')
plt.show()
```

## Create a Histogram

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Price range'], bins=4, kde=False, color='skyblue', edgecolor='black')

plt.title('Histogram of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.xticks([1, 2, 3, 4])
plt.show()
```

## Calculate the percentage of restaurants in each price range

```python
total_restaurants = len(data)
price_percentages = (price_counts / total_restaurants * 100).round(2)

# Create a DataFrame for clear viewing
price_summary = pd.DataFrame({
    'Price Range': price_counts.index,
    'Number of Restaurants': price_counts.values,
    'Percentage': price_percentages.values
})

print(price_summary)
```
